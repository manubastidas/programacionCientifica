###  **2.Configuración global**

Si usamos `git config --global` estamos diciendole a git que lo configuraremos globalmente y no por proyecto

+ `git config --global user.name "nombre"`
Asigna el nombre

+ `git config --global user.email correoelectronico` Asignamos un correo electrónico a git

+ `git config --global core.editor "code --wait" ` configura el editor de texto por defecto (VScode, jetbrains, etc). La opción de **--wait** significa que la terminal se quedará esperando hasta que cerremos el editor de texto

**Ejm**

+ Escribes **git commit** en la terminal.

+ Git necesita que escribas un mensaje (ej: "Cambio1").

+ **Si NO usas --wait**: Git abre tu editor, pero como la terminal no se detiene, Git piensa: "Bueno, ya abrí el editor, asumo que ya terminó". **Intenta hacer el commit antes de que tú alcances a escribir una sola letra.** El commit falla porque el mensaje está vacío.

+ **Si USAS --wait**: Git abre PyCharm y se queda esperando. **Tú escribes tu mensaje con calma, guardas, cierras la pestaña** y, en ese preciso instante, **Git detecta que terminaste, lee lo que escribiste y completa el commit.**

(``git config --global core.autocrlf true``)

Hay 2 personas trabajando uno en windows otro en linux-mac.
+ En windows el salto de línea se hace con dos caracteres especiales: `Carriage return(CR)` y `Line feed(LF)`
+ En linux-mac se hace solo con uno: `Line Feed(LF)`

El de windows al querer subir código va a tener que eliminar el `CR`, y por el contrario si desea descargar código del repositorio debe agregar `CR`. Para esto `core.autocrlf` debe tener el valor **True**

Para linux-mac, **al subir** ellos no agregan el caracter CR, por lo que git no debería hacer ninguna acción. **Sin embargo** si por alguna razón se agrega git; lo debe de eliminar. Para esto `core.autocrlf` debe tener el valor **input**

### **3.Flujo de trabajo**

Nosotros empezamos trabajando en el pc, podemos agregar y modificar cualquier archivo. No se agregan automáticamente al repositorio.

Para agregarlos ejecutamos `git add`. Al ejecutarlo seleccionamos los archivos que queramos, y los pasamos a una etapa llamada stage. **Acá verificamos los cambios que eventualmente queremos pasar al repositorio**

En stage puede que hayan algunos cambios que no queremos que pasen al repositorio, por lo que se pueden sacar.

**Para pasar todos los cambios al repositorio se usa** `git commit`, que pasa los archivos que estaban en la etapa stage, y los pasa a la etapa commit.

**Para subir los cambios en la etapa de commit al servidor**, se usa `git push`


Si por ejemplo en el computador eliminamos un archivo, igual debemos ejecutar `git add` contra el archivo que estamos eliminando. Así, en stage entra un archivo que hemos eliminado(Así manetenmos el registro de los cambios). Seguido de esto podemos seguir el flujo, commit, push, que terminará eliminando ese archivo del repositorio.




### **4. Comandos**

+ `ls` Lista todos los archivos y carpetas dentro de un directorio
+ `pwd` Se usa para saber en qué carpeta me encuentro actualmente
+ `cd carpeta` Para movernos entre las distintas carpetas. Para devolvernos de un directorio `cd ..` devuelve uno más atrás en la jerarquía
+ `mkdir nombreCarpeta` Para crear un nuevo directorio-carpeta
---
+ `git init` Inicializa un repositorio en la carpeta que estamos parados.

  Ver que si hacemos `ls` dentro de la carpeta que inicializamos no se ve el directorio `/.git`

  `ls -a` sirve para ver **Todo** el contenido, incluido los directorios ocultos (como /.git)

  Si navegamos a `.git` y hacemos `ls` aparecen todos los archivos que se usan en git para gestionar los proyectos (**Detalles de implementación**)

  La carpeta .git no se comparte al subir o descargar entre desarrolladores.

+ `git status` Muestra el estado actual del repositorio. Los `untracked files` son archivos que git no está subiendo. Por defecto git no sigue a todos los archivos dentro del proyecto, nosotros debemos seleccionarlos.

+ `git add nombre_archivo` Selecciona los untracked files y los monta a la etapa stage. Si ponemos en nombre_archivo un punto `git add .` estamos agregando todos los archivos(Mala práctica). **Para agregar varios archivos al mismo tiempo** `git add archivo1 archivo 2`

  **¿Qué ocurre si modificamos archivo 2 cuando está en stage?**
  Los cambios que ya estaban en stage quedarán en stage, los cambios nuevos se deben volver a agregar a stage con `git add`

  Recordar que a stage estamos pasando son los **cambios que hemos hecho** no los archivos mismos

+ `git commit` Commit significa comprometer. Para poder hacer commit hay 2 formas: `git commit -m "mensaje"`.  

  `git commit` sin mensaje abrirá el editor y pedirá un mensaje para agregar al commit

+  `rm nombrea_archivo` Sirve para crear la eliminación de un archivo que no se encuentra en la etapa de stage. Después de esto se debe agregar a stage con `git add nombrea_archivo`

    **Git tiene una abreviación de este paso, que agrega de una vez a stage la eliminación es: `git rm nombre_archivo`**

+ `git restore --staged nombre_archivo` Para deshacer algún cambio que montamos a la etapa de stage. Sin embargo si hacemos `ls` el archivo sigue eliminado. Para eso:


+ `git restore nombre_archivo` Descarta los cambios que hicimos con el archivo después de haberlo sacado del stage
---

+ `mv nombre_archivo nuevo_nombre_archivo` Esto cambia el nombre del archivo. "Copia lo que hay en el primero en el segundo". Para que git los siga, hay que agregar los cambios de ambos: `git add nombre_archivo nuevo_nombre_archivo` así se monta a stage el cambio de nombre.

    **Igual que para borrar existe la abreviación de estos comandos con `git mv nombre_archivo nuevo_nombre_archivo` que montan inmediatamente la acción a stage**

**A veces queremos archivos de configuración que sean específicos solo para nuestra máquina**.

**Ver los cambios**

Al cambiar un archivo podemos ver qué fue lo que cambiamos con:

+ `git diff` Ver que hay unas convenciones:
  1. @@num1, num2, num3@@ : ejem
  ```
  @@-1 +1,3@@
  - linea de texto original (se borro)
  \No newline at end of file (El archivo es solo 1 línea de texto)
  + linea de texto original
  + linea de texto adicional 1
  + linea de texto adicional 2
  \No newline at end of file (fin del archivo)


  ```
  Qué pasó? Git eliminó la línea de texto original y la volvió a pegar porque no habíamos presionado enter, no existía un salto de línea(En python se vería así "linea original"). Por lo que agrega la línea de texto original("línea de texto original\n"), pero añadiendo el salto de línea.

  Los números:
  + -1 indica que hay un cambio en la línea de código 1 de la versión anterior
  + +1 Indica que en la versión siguiente hay un cambio que comienza en la línea 1
  + 3 Significa los números de cambios que ha extraído

  **Para salir de diff se presiona la letra q**

  `git diff --staged` Podemos ver los cambios que se encuentran en la etapa de staged
