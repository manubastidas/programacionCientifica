# Curvas de Lissajous
### Programación Científica 2026-1 · Universidad Nacional de Colombia

Una curva de Lissajous describe el movimiento de un punto sometido a dos oscilaciones armónicas perpendiculares:

$$x(t) = A \sin(a t + \delta)$$
$$y(t) = B \sin(b t)$$

La forma de la curva depende de la relación $a/b$ y del desfase $\delta$.

```
import numpy as np
import matplotlib.pyplot as plt
```

```
# ── PARÁMETROS — modifica estos valores ──────────────────────────────────────

a     = 5.2          # frecuencia en x  (entero positivo)
b     = 2.7          # frecuencia en y  (entero positivo)
delta = np.pi / 2  # desfase en radianes  (ej: np.pi/6, np.pi/3, np.pi/2)

A     = 1.0        # amplitud en x
B     = 1.0        # amplitud en y

COLOR = "#7A5CFF"  # color de la curva  (código hex)
FONDO = "#0d1117"  # color de fondo     (código hex)

GROSOR = 2.5       # grosor de la línea
ALPHA  = 0.7      # transparencia de la curva  (0.0 – 1.0)

# ─────────────────────────────────────────────────────────────────────────────
```

```
t = np.linspace(0, 2 * np.pi, 4000)
x = A * np.sin(a * t + delta)
y = B * np.sin(b * t)

fig, ax = plt.subplots(figsize=(5, 5), dpi=150)
fig.patch.set_facecolor(FONDO)
ax.set_facecolor(FONDO)

ax.plot(x, y, color=COLOR, lw=GROSOR, alpha=ALPHA, solid_capstyle="round")

ax.set_aspect("equal")
ax.axis("off")

fig.tight_layout(pad=0.3)
plt.show()
```

 t es una lista de 4000 números linealmente separados entre 0 y 2*pi
 x y y son dos funciones sinosoidales que representan el lazo en los t valores
