## Imagen 3


````python
lat = np.linspace(-90, 90, 180)          # 180 valores de latitud: de -90° a 90°
lon = np.linspace(-180, 180, 360)        # 360 valores de longitud: de -180° a 180°
LON, LAT = np.meshgrid(lon, lat)         # malla 2D de coordenadas
rng = np.random.default_rng(42)          # generador de números aleatorios reproducible
T = (30 * np.cos(np.radians(LAT))        # componente latitudinal: máx en ecuador
     - 20                                # desplazamiento global (temperatura base)
     + 5 * np.sin(np.radians(2 * LON))   # variación longitudinal suave
     + 3 * rng.standard_normal((180, 360)))  # ruido gaussiano (variabilidad local)

print(T) # muestra la matriz de los datos
print(T.shape) # muestra la forma de la matriz

plt.imshow(T)   # visualización sin argumentos
#plt.show()      # muestra la figura
````

### 🔍 Exploración

```python
print(T.shape)             # ¿qué representa cada dimensión?
print(T.dtype)             # ¿es entero o flotante? ¿por qué?
print(T.min(), T.max())    # ¿cuál es el rango físico de temperaturas?
print(T.mean())            # ¿cuál es la temperatura media global?
print((T > 0).sum())       # ¿cuántos píxeles tienen temperatura positiva?
```
