* **Explicación técnica:** Un servidor es un computador remoto al que nos podemos conectar para hacer algunas tareas. En este caso se usa SSH (Secure Shell) para conectarse. Esto es un protocolo que cifra la comunicación y proporciona acceso interactivo seguro a la línea de comandos de esa máquina. Además, se nos pide especificar el puerto 2220 en nuestro comando (-p 2220); esto es un identificador numérico funciona como la puerta de red específica asignada por los administradores de Bandit para dirigir nuestro tráfico al juego y aislarlo del puerto SSH predeterminado (22).
