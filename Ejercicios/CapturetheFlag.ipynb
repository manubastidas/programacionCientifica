# Actividad Colaborativa: Capture the Flag (Bandit)

**Objetivo:** Practicar conexiones SSH y comandos de terminal resolviendo el wargame Bandit https://overthewire.org/wargames/bandit/.
Documentar las soluciones en equipo utilizando un flujo estricto de GitHub.

**Método:**

1. **Crea una rama:** Crea una rama nueva para tu nivel. Usa `git checkout -b bandit-nivel-X` (reemplaza la X con tu nivel).

2. **Captura la bandera:** Accede al servidor de Bandit y encuentra la solución del siguiente nivel usando comandos de Linux en la terminal remota.

3. **Edita el archivo:** En este archivo `CapturetheFlag.ipynb` añade tu solución siguiendo la plantilla y las instrucciones.

```markdown
### Nivel X -> Nivel X+1

* **Autor:** [Tu Nombre y Apellido]
* **Comandos de solución:**
* **Explicación técnica:** [Explica la función de cada comando utilizado y cómo extrajiste la bandera]
* **Próximo nivel:** @[Usuario de GitHub del siguiente estudiante]
```

4. **Sube los cambios y abre un Pull Request (PR):** Abre un PR desde tu rama hacia `main`. Designa explícitamente al compañero que debe resolver el siguiente nivel.

5. **Asigna revisores:** Etiqueta y asigna a dos compañeros diferentes (y la profe) como revisores de tu PR. Los dos revisores verifican la sintaxis del comando y el formato, y aprueban los cambios.
