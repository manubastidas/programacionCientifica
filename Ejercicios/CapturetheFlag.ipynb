* **Explicación técnica:**

En este nivel, la contraseña se encuentra en el único archivo legible por humanos dentro de la carpeta `inhere`. Dado que la mayoría eran archivos de datos binarios (ilegibles), busqué el que decía "ASCII text". El prefijo `./` es necesario porque los nombres de los archivos comienzan con un guion (`-`), y sin él, el comando los confundiría con opciones del propio comando.

* **Explicación técnica:** 
Los primeros dos comandos, como han mostrado los anteriores compañerxs, son para conectarse correctamente al servidor. Ahora bien desglocemos el comando: `find / -user bandit7 -group bandit6 -size 33c`.  
`find /` busca en todo el disco partiendo desde raíz, y a continuación se le pueden poner banderas (flags) para indicar los filtros con los cuales debe buscar.  
Aquí viene el primer filtro que es `-user bandit7` el cual reduce la búsqueda a ese usuario en específico.  
Luego `-group bandit6` le indica que debe buscar solo los archivos pertenecientes al grupo bandit6.  
Finalmente `-size 33c` delimita la búsqueda a archivos que solo pesan 33 bytes.  
Este comando nos arroja una lista de resultados, en la cual está la contraseña, solo debemos ignorar los resultados en los cuales no tenemos acceso para buscar (donde muestra  `Permission denied`).  
Ya con la direccion donde esta la contraseña podemos usar finalmente `cat /var/lib/dpkg/info/bandit7.password` para acceder y obtener la contraseña. Y como último paso ejecutamos `exit` para salir.
